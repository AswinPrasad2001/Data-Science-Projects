```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold_data = pd.read_csv('/content/gld_price_data.csv')
```

```python
correlation = gold_data.corr(numeric_only=True)
```

```python
from ast import If
plt.figure(figsize=(8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
print(correlation['GLD'])
```

```python
sns.displot(gold_data['GLD'],color='green')
```

```python
X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=2)
```

```python
regression = RandomForestRegressor(n_estimators=100)
```

```python
regression.fit(X_train,Y_train)
```

```python
test_data_predicition = regression.predict(X_test)
```

```python
print(test_data_predicition)
```

```python
error_score = metrics.r2_score(Y_test, test_data_predicition)
print("R square error: ", error_score)
```

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label = 'Actual Value')
plt.plot(test_data_predicition, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
